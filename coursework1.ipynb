This is my own function called fast_matrix_product.

```python
import numpy as np


def fast_matrix_product(mat1, mat2):
    assert mat1.shape[1] == mat2.shape[0]
    result = []
    for r in range(mat1.shape[0]):
        column = []
        for c in range(mat2.shape[1]):
            value = np.dot(mat1[r, :], mat2[:, c])
            column.append(value)
        result.append(column)
    return np.array(result)


matrix1 = np.random.rand(3, 3)
matrix2 = np.random.rand(3, 3)

print(fast_matrix_product(matrix1, matrix2))
print(matrix1 @ matrix2)
```

Reasons
1.This function decreases a for loop, so this may speed up the arithmetic.
2.By putting for loop about c in for loop about r, this function amend the order to a better one.
3.By using np.dot, this funtion increases the speed and saves time.

```python
import numpy as np


def slow_matrix_product(mat1, mat2):
    """Multiply two matrices."""
    assert mat1.shape[1] == mat2.shape[0]
    result = []
    for c in range(mat2.shape[1]):
        column = []
        for r in range(mat1.shape[0]):
            value = 0
            for i in range(mat1.shape[1]):
                value += mat1[r, i] * mat2[i, c]
            column.append(value)
        result.append(column)
    return np.array(result).transpose()


matrix1 = np.random.rand(10, 10)
matrix2 = np.random.rand(10, 10)

print(slow_matrix_product(matrix1, matrix2))
print(matrix1 @ matrix2)
```

```python
matrix_size = [2, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

slow_times = []
fast_times = []

for n in matrix_size:
  test_mat1 = np.random.rand(n,n)
  test_mat2 = np.random.rand(n,n)
  t_slow = %timeit -o -q slow_matrix_product(test_mat1, test_mat2)
  t_fast = %timeit -o -q fast_matrix_product(test_mat1, test_mat2)
  slow_times.append(t_slow.average)
  fast_times.append(t_fast.average)
  print(f"[Slow] [N={n}] Average Run time: {t_slow.average}")
  print(f"[Fast] [N={n}] Average Run Time: {t_fast.average}")
```

```python
import matplotlib.pylab as plt
import numpy as np
plt.loglog(matrix_size, slow_times)
plt.loglog(matrix_size, fast_times)



plt.xlabel("Matrix sizes")
plt.ylabel("Time")
plt.title("Matrix mattric product")
plt.savefig("myplot.pdf")
```

```python
import numpy as np
import numba
from numba import jit

@jit
def fast_matrix_product_jit(mat1, mat2):
    assert mat1.shape[1] == mat2.shape[0]
    result = np.zeros((mat1.shape[0], mat2.shape[1]), dtype=mat1.dtype)

    for r in range(mat1.shape[0]):
        for c in range(mat2.shape[1]):
            result[r, c] = np.dot(mat1[r, :], mat2[:, c])

    return result
matrix1 = np.random.rand(3, 3)
matrix2 = np.random.rand(3, 3)

print(fast_matrix_product(matrix1, matrix2))
print(matrix1 @ matrix2)

```

```python
import matplotlib.pylab as plt
import numpy as np

matrix_size = [2, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

fast_times_jit = []
fast_times = []
numpy_time = []

for n in matrix_size:
  test_mat1 = np.random.rand(n,n)
  test_mat2 = np.random.rand(n,n)
  t_fast_jit = %timeit -o -q fast_matrix_product_jit(test_mat1, test_mat2)
  t_fast = %timeit -o -q fast_matrix_product(test_mat1, test_mat2)
  t_numpy = %timeit -o -q test_mat1 @ test_mat2
  fast_times_jit.append(t_fast_jit.average)
  fast_times.append(t_fast.average)
  numpy_time.append(t_numpy.average)




plt.loglog(matrix_size, fast_times_jit)
plt.loglog(matrix_size, fast_times)
plt.loglog(matrix_size, numpy_time)

plt.plot(matrix_size, fast_times_jit, 'o-', label="fast_times_jit")
plt.plot(matrix_size, fast_times, 'o-', label="fast_times")
plt.plot(matrix_size, numpy_time, 'o-', label="numpy_time")
plt.legend()

plt.xlabel("Matrix sizes")
plt.ylabel("Time")
plt.title("Matrix mattric product")
plt.savefig("myplot.pdf")
```

```python
import numpy as np
import matplotlib.pylab as plt
matrix_size = [2, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

fast_times_jit_cc = []
fast_times_jit_cf = []
fast_times_jit_fc = []
fast_times_jit_ff = []

for n in matrix_size:
  test_mat1_c = np.random.rand(n,n)
  test_mat2_c = np.random.rand(n,n)
  test_mat1_f = np.asfortranarray(test_mat1_c)
  test_mat2_f = np.asfortranarray(test_mat2_c)

  t_fast_jit_cc = %timeit -o -q fast_matrix_product_jit(test_mat1_c, test_mat2_c)
  t_fast_jit_cf = %timeit -o -q fast_matrix_product_jit(test_mat1_c, test_mat2_f)
  t_fast_jit_fc = %timeit -o -q fast_matrix_product_jit(test_mat1_f, test_mat2_c)
  t_fast_jit_ff = %timeit -o -q fast_matrix_product_jit(test_mat1_f, test_mat2_f)
  fast_times_jit_cc.append(t_fast_jit_cc.average)
  fast_times_jit_cf.append(t_fast_jit_cf.average)
  fast_times_jit_fc.append(t_fast_jit_fc.average)
  fast_times_jit_ff.append(t_fast_jit_ff.average)




plt.loglog(matrix_size, fast_times_jit_cc)
plt.loglog(matrix_size, fast_times_jit_cf)
plt.loglog(matrix_size, fast_times_jit_fc)
plt.loglog(matrix_size, fast_times_jit_ff)

plt.plot(matrix_size, fast_times_jit_cc, 'o-', label="fast_times_jit_cc")
plt.plot(matrix_size, fast_times_jit_cf, 'o-', label="fast_times_jit_cf")
plt.plot(matrix_size, fast_times_jit_fc, 'o-', label="fast_times_jit_fc")
plt.plot(matrix_size, fast_times_jit_ff, 'o-', label="fast_times_jit_ff")
plt.legend()

plt.xlabel("Matrix sizes")
plt.ylabel("Time")
plt.title("Matrix mattric product")
plt.savefig("myplot.pdf")







```
